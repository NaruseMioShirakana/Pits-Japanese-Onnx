```python
import models_onnx
import IPython.display as ipd
import utils
from text.symbols import symbol_len, lang_to_dict
```

```python
hps = utils.get_hparams_from_file("./configs/config_en.yaml")
device = "cpu"
net_g = models_onnx.SynthesizerTrn(symbol_len(hps.data.languages),
                            hps.data.filter_length // 2 + 1,
                            hps.train.segment_size //
                            hps.data.hop_length,
                            midi_start=-5,
                            midi_end=75,
                            octave_range=24,
                            n_speakers=len(hps.data.speakers),
                            **hps.model).to(device)
_ = net_g.eval()
_ = utils.load_checkpoint("./logs/pits_vctk_AD_3000.pth", model_g=net_g)
```

```python
audio = net_g.export_onnx(True).squeeze(0).detach().numpy()
ipd.display(ipd.Audio(audio, rate=22050, normalize=False))
```
